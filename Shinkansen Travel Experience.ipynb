## Converting variables into categorical datatype

```python
df_train = df_train.astype({'Gender': 'category', 'CustomerType': 'category', 'TypeTravel': 'category', 'Travel_Class': 'category', 'Seat_comfort': 'category', 'Seat_Class': 'category', 'Arrival_time_convenient': 'category', 'Catering': 'category', 'Platform_location': 'category', 'Onboardwifi_service': 'category', 'Onboard_entertainment': 'category', 'Online_support': 'category', 'Onlinebooking_Ease': 'category', 'Onboard_service': 'category', 'Leg_room': 'category', 'Baggage_handling': 'category', 'Checkin_service': 'category', 'Cleanliness': 'category', 'Online_boarding': 'category'}) 
df_test = df_test.astype({'Gender': 'category', 'CustomerType': 'category', 'TypeTravel': 'category', 'Travel_Class': 'category', 'Seat_comfort': 'category', 'Seat_Class': 'category', 'Arrival_time_convenient': 'category', 'Catering': 'category', 'Platform_location': 'category', 'Onboardwifi_service': 'category', 'Onboard_entertainment': 'category', 'Online_support': 'category', 'Onlinebooking_Ease': 'category', 'Onboard_service': 'category', 'Leg_room': 'category', 'Baggage_handling': 'category', 'Checkin_service': 'category', 'Cleanliness': 'category', 'Online_boarding': 'category'})
```

```python
for i in df_train.columns: 
    if df_train[i].dtype =='category': 
        print('\n')
        print('feature:',i)
        print(pd.Categorical(df_train[i].unique()))
        print(pd.Categorical(df_train[i].unique()).codes)
        df_train[i] =pd.Categorical(df_train[i]).codes
        
for i in df_test.columns: 
    if df_test[i].dtype =='category': 
        print('\n')
        print('feature:',i)
        print(pd.Categorical(df_test[i].unique()))
        print(pd.Categorical(df_test[i].unique()).codes)
        df_test[i] =pd.Categorical(df_test[i]).codes
```

## Dropping unnecessary columns

```python
df_train.drop('ID',axis = 1,inplace = True)
test_file_ids = df_test['ID'] ## necessary step
df_test.drop('ID',axis = 1,inplace = True)
```

## Splitting training dataset into train and validation set

```python
X = df_train.drop(['Overall_Experience'],axis = 1)
y = df_train['Overall_Experience']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_validation, y_train, y_validation = train_test_split(X,y,test_size = 0.2,random_state = 0)
```

## Model training - Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(n_estimators=10,criterion='gini',max_depth=10,random_state=0)
rf_model.fit(X_train,y_train)

y_pred_train = rf_model.predict(X_train)
y_pred_val = rf_model.predict(X_validation)
```

```python
from sklearn import metrics
dt_acc_train = metrics.accuracy_score(y_train,y_pred_train)
dt_acc_val = metrics.accuracy_score(y_validation,y_pred_val)
print("Train Accuracy:",dt_acc_train)
print("Validation Accuracy:",dt_acc_val)
print()

dt_roc_train = metrics.roc_auc_score(y_train,y_pred_train)
dt_roc_val = metrics.roc_auc_score(y_validation,y_pred_val)
print("Train ROC AUC score:",dt_roc_train)
print("Validation ROC AUC score:",dt_roc_val)
print()

dt_clr_train = metrics.classification_report(y_train,y_pred_train)
dt_clr_val = metrics.classification_report(y_validation,y_pred_val)
print("Train Classification Report:\n",dt_clr_train)
print("Validation Classification Report:\n",dt_clr_val)
```

## Model Training - XG Boost

### Ensembling RF Model and XG Boost Model with best params

```python
from xgboost import XGBClassifier
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV

# Define the parameter grid
param_grid = {'learning_rate': [0.1, 0.2, 0.5], 'max_depth': [5, 8, 10], 'n_estimators': [200, 300, 500]}

# Create an instance of the XGBoost model
xgb = XGBClassifier()

# Use RandomizedSearchCV or GridSearchCV to fine-tune the hyperparameters
xgb_cv = RandomizedSearchCV(xgb, param_grid, cv=5)

# Fit the model to the training data
xgb_cv.fit(X_train, y_train)

# Make predictions on the validation set
y_pred = xgb_cv.predict(X_validation)

# Evaluate the model's performance
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
print("Accuracy:", accuracy_score(y_validation, y_pred))
print("F1 Score:", f1_score(y_validation, y_pred, average='weighted'))
print("Precision:", precision_score(y_validation, y_pred, average='weighted'))
print("Recall:", recall_score(y_validation, y_pred, average='weighted'))
```

## Predicting on test data using ensembled model of RF and XG Boost and creating a submission file

```python
final_predictions = xgb_cv.predict(df_test)
submission=pd.DataFrame([test_file_ids,final_predictions]).T
submission.rename(columns={"ID": "ID", "Unnamed 0": "Overall_Experience"},inplace=True)
submission.to_csv(r'F:\Personal\PGP\Hackathon\submission.csv',index = False)
```
